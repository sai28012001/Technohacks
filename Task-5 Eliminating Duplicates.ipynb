Identify and remove duplicate values in an iris dataset

## Importing the python libraries

```python
import pandas as pd
```

## Importing the dataset

```python
data=pd.read_csv("C:/Users/saiba/Desktop/Online Internships/Technohacks/Task-5/Iris.csv")
data
```

## checking duplicates of entire rows

```python
data.duplicated()
```

```python
data.duplicated().value_counts()
```

## Removing the rows which are completely duplicated

```python
data.drop_duplicates()
```

## Removing duplicates where the values are duplicated in one column

```python
data.drop_duplicates("Species")
```

```python
data.duplicated(subset=['SepalLengthCm']).value_counts()
```

# Removing duplicates based on more than one column

```python
data.drop_duplicates(subset=['SepalLengthCm','SepalWidthCm'])
```

## Removing duplicates based on more than one column and keep first/last

```python
data.drop_duplicates(subset=['SepalLengthCm','SepalWidthCm'],keep="last")
```
